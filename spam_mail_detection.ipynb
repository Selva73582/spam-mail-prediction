```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
df=pd.read_csv('email.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
label_encoder=LabelEncoder()
df['target']=label_encoder.fit_transform(df['Category'])
df.head()
```

# Detail of category
## Ham - 0
## Spam - 1


```python
df=df.drop(columns=['Category'],axis='columns')
df.head()
```

# Text preprocessing


1. Lowercasing
2. Removing Punctuation
3. Tokenization
4. Stopword Removal
5. Stemming/Lemmatization

```python
from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
X_train,X_test, y_train, y_test=train_test_split(df['Message'],df['target'],test_size=0.3,random_state=42)
```

```python
vectorizer=TfidfVectorizer(lowercase=True,stop_words=None,min_df=1)
```

```python
X_train_transfered=vectorizer.fit_transform(X_train)
X_test_transfered=vectorizer.transform(X_test)
```

```python
model=SVC(kernel='rbf')
```

```python
model.fit(X_train_transfered,y_train)
```

```python
y_pred = model.predict(X_test_transfered)

accuracy_score(y_pred,y_test)
```

```python
import joblib

joblib.dump(model,'logistic.pkl')
joblib.dump(vectorizer,'transformModel.pkl')
```
